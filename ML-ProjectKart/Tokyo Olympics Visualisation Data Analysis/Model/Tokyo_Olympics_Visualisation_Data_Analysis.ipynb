# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the  first Dataset

```python
df_athletes=pd.read_excel("/content/drive/MyDrive/Tokyo Olympics Dataset/Athletes.xlsx")
df_athletes
```

```python
df_athletes['Discipline'].unique()
```

```python
df_athletes['NOC'].unique()
```

```python
df_athletes.isnull().sum()
```

```python
NOC_counts=df_athletes.NOC.value_counts()
NOC_counts
```

```python
Discipline_counts=df_athletes.Discipline.value_counts()
Discipline_counts
```

# Percentage distribution of different sports using pie-chart

```python
plt.figure(figsize=(30,10))
plt.pie(Discipline_counts,labels=Discipline_counts.index,autopct='%1.2f%%',startangle=150,shadow=True)
```

Athletics has more participation as compared to other sports

## Country-Wise participation using piechart

```python
plt.figure(figsize=(30,10))
plt.pie(NOC_counts,labels=NOC_counts.index,autopct='%1.2f%%',startangle=150,shadow=True)
```

United States has more players participation followed by Japan ,Australia

## Top 10 participating countries using bar plot

```python
#Top 10 countries participating
top_10_countries=df_athletes.NOC.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot for the top 10 countries
plt.figure(figsize=(16,9))
plt.xticks(rotation=90)
sns.barplot(x=top_10_countries.index,y=top_10_countries,palette='PuOr')
```

## Top 10 popular sports of Olympics using bar plot

```python
top_10_sports=df_athletes.Discipline.value_counts().sort_values(ascending=False).head(10)
top_10_sports
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation=90)
sns.barplot(x=top_10_sports.index,y=top_10_sports,palette='coolwarm')
```
